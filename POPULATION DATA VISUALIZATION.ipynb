```python
import matplotlib.pyplot as plt
import numpy as np

# Creating a random dataset for demonstration: ages in a population
np.random.seed(42)  # For reproducibility
ages = np.random.randint(18, 80, size=100)  # 100 random ages between 18 and 80

# Plotting a histogram to visualize the distribution of ages
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=12, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Ages in a Population')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```
